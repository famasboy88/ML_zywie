```python
cred = credentials.Certificate('service.json')
firebase_admin.initialize_app(cred, {
    'databaseURL' : 'https://zywie-2b7c2.firebaseio.com'
})
rootDB = db.reference().child("food_title")
root = rootDB.get()
```

```python
from nltk.corpus import stopwords

cachedStopWords = stopwords.words("english")
print(cachedStopWords)
```

```python
# empty array
array_recipe = []

# regular expressions to clean invalid characters and number
pattern = re.compile('[^A-Za-z ]+')
# regular expressions to clean stop words
stopwordsReg = re.compile(r'\b(' + r'|'.join(cachedStopWords) + r')\b\s*')

for cat, arrItem in root.items():
        recipe_lists = arrItem.get("list_recipe")
        if(recipe_lists is not None):
            recipe_var=''
            for recipe in recipe_lists:
                recipe = recipe.lower()
                recipe = stopwordsReg.sub('', recipe)
                recipe_filter = pattern.sub('', recipe)
                recipe_filter = re.sub(' +', ' ', recipe_filter)
                recipe_filter = re.sub('^[^\w]*', '', recipe_filter)
                recipe_filter.strip()
                recipe_var += recipe_filter+"|"
#open file
            f = open('recipe\\'+pattern.sub('', cat)+'.txt', 'w')
            f.write(recipe_var)
            f.close()
            print(recipe_var)
            
```
